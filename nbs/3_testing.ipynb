```python
df_train = pd.read_csv('../data/data_clean/train.csv')
df_val = pd.read_csv('../data/data_clean/val.csv')
data_clas = TextClasDataBunch.from_df('../data/data_clean', df_train, df_val, text_cols=0, label_cols=1, bs=32)
learn = text_classifier_learner(data_clas, arch=AWD_LSTM, pretrained=True, drop_mult=0.3, model_dir='.')
learn.load_encoder('models/twitter-sentiment-enc', device=torch.device('cpu'))
learn.load('models/twitter-sentiment-classifier')
```
