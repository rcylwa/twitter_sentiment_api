```python
#from fastai import *
import pandas as pd
from fastai.text import *
```

Loading previously saved model:

```python
df_train = pd.read_csv('../data/data_clean/train.csv')
df_val = pd.read_csv('../data/data_clean/val.csv')
data_clas = TextClasDataBunch.from_df('../data/data_clean', df_train, df_val, text_cols=0, label_cols=1, bs=32)
learn = text_classifier_learner(data_clas, arch=AWD_LSTM, pretrained=True, drop_mult=0.3, model_dir='.')
learn.load_encoder('models/twitter-sentiment-enc')
learn.load('models/twitter-sentiment-classifier')
```

Loading the test set:

```python
df_test = pd.read_csv('../data/data_clean/test.csv')
df_test.head()
```

```python
preds = [str(learn.predict(x)[0]) for x in df_test.text]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(df_test.airline_sentiment, preds)
```

We get an accuracy score similar to train and validation sets, which is a good sign that the model generalizes well.
